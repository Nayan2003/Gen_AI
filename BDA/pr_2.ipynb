```python
import pyspark
from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder.appName("WordCount").getOrCreate()

# Read input text file (Replace 'input.txt' with your file path)
rdd = spark.sparkContext.textFile("/content/inputText.txt")  # give path of data file

# Map: Split lines into words and create (word, 1) pairs
words = rdd.flatMap(lambda line: line.split(" ")) \
           .map(lambda word: (word, 1))

# Reduce: Count occurrences of each word
word_counts = words.reduceByKey(lambda a, b: a + b)

# Collect and display results
for word, count in word_counts.collect():
    print(f"{word}: {count}")

# Stop Spark Session
spark.stop()
```

# 🔹 Python List Methods

## ✅ 1. `append()`

👉 End me element add karta hai

```python
numbers = [1, 2, 3]
numbers.append(4)

print(numbers)  # [1, 2, 3, 4]
```

---

## ✅ 2. `insert()`

👉 Specific position pe element add karta hai

```python
numbers = [1, 2, 3]
numbers.insert(1, 10)

print(numbers)  # [1, 10, 2, 3]
```

---

## ✅ 3. `extend()`

👉 Dusri list ke elements add karta hai

```python
a = [1, 2]
b = [3, 4]

a.extend(b)
print(a)  # [1, 2, 3, 4]
```

---

## ✅ 4. `remove()`

👉 First matching value delete karta hai

```python
numbers = [1, 2, 3, 2]
numbers.remove(2)

print(numbers)  # [1, 3, 2]
```

❌ Error agar value exist nahi karti:

```python
numbers.remove(10)  # ValueError
```

---

## ✅ 5. `pop()`

👉 Index se element remove karta hai (default last)

```python
numbers = [1, 2, 3]
numbers.pop()

print(numbers)  # [1, 2]
```

👉 Specific index:

```python
numbers.pop(0)  # remove first element
```

---

## ✅ 6. `clear()`

👉 Puri list empty kar deta hai

```python
numbers = [1, 2, 3]
numbers.clear()

print(numbers)  # []
```

---

## ✅ 7. `index()`

👉 Value ka position batata hai

```python
numbers = [10, 20, 30]
print(numbers.index(20))  # 1
```

❌ Error if not found

---

## ✅ 8. `count()`

👉 Kitni baar element aaya hai

```python
numbers = [1, 2, 2, 3]
print(numbers.count(2))  # 2
```

---

## ✅ 9. `sort()`

👉 List ko sort karta hai

```python
numbers = [3, 1, 2]
numbers.sort()

print(numbers)  # [1, 2, 3]
```

👉 Reverse:

```python
numbers.sort(reverse=True)
```

---

## ✅ 10. `reverse()`

👉 List ko ulta kar deta hai

```python
numbers = [1, 2, 3]
numbers.reverse()

print(numbers)  # [3, 2, 1]
```

---

## ✅ 11. `copy()`

👉 List ki copy banata hai

```python
a = [1, 2, 3]
b = a.copy()

print(b)
```

## enumerate() --> it will return the index and the item in the list

##### Enumerate --> helps us with getting the index value along with the actual in a combination that we are storing inside the list

````python
## enumerate() --> it will return the index and the item in the list
fruits = ["mango", "banana", "grapes", "orange", "kiwi","pineappele", "watermelon", "papaya", "strawberry", "blueberry"]
````

````python
for i in enumerate(fruits):
    print(i)
````

````python
for index, value in enumerate(fruits): # default start index is 0
    print(index, value)
````

````python
for index, value in enumerate(fruits,start=1): # giving the start index as 1 but default start index is 0
    print(index, value)
````

## List Comprehension

#### we are goin to use looping statements inside the list so that we can store the values inside the list in an oterable format and in list comprehension conditional statements are allowed as well 

## Syntax for list comprehension is 

### [expression for item in iterable if condition]

### expression --> What we want to print

### for item in iterable -- > looping through the iterable Looping statement

### if condition --> Conditional statement

````python
# i need to store values from 1 to 50 inside the list




list_items = [x for x in range(1,51)]

print(list_items)
````

````python
list_sqares = [x**2 for x in range(1,51)]
print(list_sqares)
````

````python
# if i am only intrested in looking over the sqare values which are even


even_items = [x**2 for x in range(1,51) if x**2 % 2 == 0]
print(even_items)
````

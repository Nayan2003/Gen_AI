🧠 Python Operators – Detailed Notes
🔹 1. What are Operators?

Operators are symbols or keywords that perform operations on variables and values.

# 1. Arithmetic Operators

Basic maths ke liye use hote hain

```python


a = 10
b = 3

print("addion operator (+):", a + b)   # 13  (Addition)
print("subtraction operator (-):", a - b)   # 7   (Subtraction)
print("multiplication operator (*):", a * b)   # 30  (Multiplication)
print("division operator (/):", a / b)   # 3.33 (Division - float)
print("floor division operator (//):", a // b)  # 3   (Floor division)
print("modulus operator (%):", a % b)   # 1   (Modulus - remainder)
print("power operator (**):", a ** b)  # 1000 (Power)
```

# 2. Comparison Operators

True/False return karte hain

```python
a = 5
b = 10

print("equality operator (a==b) ==> ({} == {}):".format(a, b), a == b)  # False
print("inequality operator (a!=b) ==> ({} != {}):".format(a, b), a != b)  # True
print("greater than operator (a>b) ==> ({} > {}):".format(a, b), a > b)   # False
print("less than operator (a<b) ==> ({} < {}):".format(a, b), a < b)   # True
print("greater than or equal to operator (a>=5) ==> ({} >= 5):".format(a), a >= 5)  # True
print("less than or equal to operator (a<=3) ==> ({} <= 3):".format(a), a <= 3)  # False
```

# 3. Logical Operators

Conditions combine karne ke liye

```python
x = True
y = False

print(x and y)  # False
print(x or y)   # True
print(not x)    # False
```

# 4. Assignment Operators

Value assign ya update karne ke liye

```python
x = 5

x += 3   # x = x + 3 → 8
x -= 2   # 6
x *= 2   # 12
x /= 3   # 4.0
```

# 5. Bitwise Operators

Binary level pe kaam karte hain (advanced)

```python
a = 5   # 101
b = 3   # 011

print(a & b)  # 1  (AND)
print(a | b)  # 7  (OR)
print(a ^ b)  # 6  (XOR)
print(~a)     # -6 (NOT)
print(a << 1) # 10 (Left shift)
print(a >> 1) # 2  (Right shift)
```

# 6. Membership Operators

Check karta hai element exist karta hai ya nahi

```python
lst = [1, 2, 3]

print(2 in lst)      # True
print(5 not in lst)  # True
```

# 7. Identity Operators

Memory location compare karte hain

```python
a = [1,2]
b = [1,2]

print(a == b)  # True (value same)
print(a is b)  # False (memory different)
```

| Type       | Operators             |
| ---------- | --------------------- |
| Arithmetic | +, -, *, /, %, //, ** |
| Comparison | ==, !=, >, <, >=, <=  |
| Logical    | and, or, not          |
| Assignment | =, +=, -=, *=, /=     |
| Bitwise    | &, |, ^, ~, <<, >>    |
| Membership | in, not in            |
| Identity   | is, is not            |


# simple calculator

```python
a = int(input("Enter a first number: "))
b = int(input("Enter a second number: "))

oprator = input("Enter an operator (+, -, *, /): ")

if oprator == '+':
    print("Result: the sum of {} and {} is {} + {} = {}".format(a, b,a,b, a + b));
elif oprator == '-':
    print("Result: the difference of {} and {} is {} - {} = {}".format(a, b,a,b, a - b));
elif oprator == '*':
    print("Result: the product of {} and {} is {} * {} = {}".format(a, b,a,b, a * b));
elif oprator == '/':
    if b != 0:
        print("Result: the division of {} by {} is {} / {} = {}".format(a, b,a,b, a / b));
    else:
        print("Error: Division by zero is not allowed.");
else:
    print("Error: Invalid operator. Please enter one of +, -, *, /.");
```

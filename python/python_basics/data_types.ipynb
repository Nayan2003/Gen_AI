# 5. list

#### Ordered, mutable (change ho sakta hai)

```python
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)
```

```python
# Error
numbers = [1, 2, 3]
print(numbers[5])
# IndexError: list index out of range 
```

##### Do
Modify kar sakte ho

```python
numbers[0] = 10
```

#### Don’t
Out-of-range index access mat karo

# 6. tuple

👉 Ordered, immutable (change nahi hota)

```python
t = (1, 2, 3)
print(t[0])
```

```python
# Error
t = (1, 2, 3)
t[0] = 10
# TypeError: 'tuple' object does not support item assignment
```

#### Do
Fixed data store karne ke liye use karo

#### Don’t
Modify karne ki koshish mat karo

# 7. set

👉 Unordered, unique elements

```python
s = {1, 2, 3, 3}
print(s)  # {1, 2, 3}
```

```python
#Error
s = {1, 2, 3}
print(s[0])
# TypeError: 'set' object is not subscriptable
```

####  Do
Unique values ke liye use karo
#### Don’t
Index use mat karo (list jaisa nahi hota)

# 8. dict (Dictionary)

👉 Key-value pairs

```python
student = {
    "name": "Nayan",
    "age": 22
}
print(student["name"])
```

```python
#Error
print(student["marks"])
# KeyError: 'marks'
```

#### Do
.get() use karo safe access ke liye

```python
student.get("marks")
```

#### Don’t
Non-existing key directly access mat karo

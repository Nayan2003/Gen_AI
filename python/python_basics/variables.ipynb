# variables naming Conventions:

#### variables name must be Descriptive:
    #### Name should be meaningful and should describe the purpose of the variable.
    #### It should be easy to understand and remember.
    #### It should not be too long or too short.   
## variables naming Conventions:
    # 1. snake_case --> used for variables and functions (e.g., my_variable)
    # 2. camelCase --> used for variables and functions (e.g., myVariable)
    # 3. PascalCase --> used for classes (e.g., MyClass)  



#### They must start with a letter or an underscore and can contain letters, digits, and underscores. They cannot start with a digit and cannot be a reserved keyword in Python.

```python
1name = "nayan"  # invalid variable name, starts with a digit
```

## type conversion

```python
## convering into int to string
age = 23   
age_str = str(age)  # converting integer to string
print(age_str)  # Output: '23'
```

```python
# every type casting is not possible for example we cannot convert string to int if the string does not represent a valid integer.
name = "nayan"
name_int = int(name)  # This will raise a ValueError because "nayan" cannot be converted to an integer.
```

```python
# bute casting is possible for example we can convert string to int if the string represents a valid integer.
age_str = "23" 
age_int = int(age_str)  # This will successfully convert the string "23" to the integer 23.
print(age_int)  # Output: 23
```

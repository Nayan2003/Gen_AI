# Whethere if the ccurrent year is leap year or not

```python
year = int(input("Enter the year: "))
print("you entered year is {}".format(year))

if year % 4 ==0:
    if year % 100 ==0:
        if year % 400 ==0:
            print("{} is a leap year".format(year))
        else:
            print("{} is not a leap year".format(year))
    else:
        print("{} is a leap year".format(year))

else:
    print("{} is not a leap year".format(year))
```

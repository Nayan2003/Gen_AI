## 1. What is a Tuple?

👉 Tuple ek ordered, immutable collection hota hai Python mein.

### Ordered → elements ka order fix hota hai
### Immutable → ek baar create ho gaya, change nahi kar sakte
### Allows duplicates

```python
t = (1, 2, 3, 4)
```

| Feature     | Tuple       | List         |
| ----------- | ----------- | ------------ |
| Syntax      | ()          | []           |
| Mutability  | ❌ Immutable | ✅ Mutable    |
| Performance | Faster      | Slower       |
| Use case    | Fixed data  | Dynamic data |

## 4. Accessing Elements

```python
#✔️ Indexing
t = (10, 20, 30)
print(t[0])   # 10
```

```python
# Negative indexing
print(t[-1])  # 30
```

```python
#Slicing
print(t[0:2])  # (10, 20)
```

### 5. Tuple is Immutable ❌

```python
#5. Tuple is Immutable ❌
t = (1, 2, 3)
t[0] = 10   # ❌ Error
```

### 👉 But agar tuple ke andar list hai, wo change ho sakti hai:

```python


t = (1, [2, 3])
t[1][0] = 100
print(t)   # (1, [100, 3])
```

## 6. Tuple Operations

```python

#✔️ Concatenation
t1 = (1, 2)
t2 = (3, 4)
print(t1 + t2)
```

```python
#✔️ Repetition
print((1, 2) * 3)
```

##  7. Tuple Methods (Very Limited )

```python
# Tuples ke sirf 2 main methods hote hain:

# ✔️ count()
t = (1, 2, 2, 3)
print(t.count(2))   # 2
```

```python
#✔️ index()
print(t.index(3))   # 3rd element index
```

## 8. Tuple Packing & Unpacking

```python
#✔️ Packing
t = 1, 2, 3
```

```python
#✔️ Unpacking
a, b, c = t
print(a, b, c)
```

```python
# ✔️ Extended unpacking
a, *b = (1, 2, 3, 4)
print(a)  # 1
print(b)  # [2,3,4]
```

## Nested dictionary

### Nested dictionary will be used for keeping a dictionary inside another dictionary. It is useful for representing complex data structures where you want to group related information together.

```python
## want to store information multiple students marks and grade in a class

nested_dict ={"student1":{"name":"Alice","marks":85,"grade":"A"}
              ,"student2":{"name":"Bob","marks":78,"grade":"B"}
              ,"student3":{"name":"Charlie","marks":92,"grade":"A+"}
              ,"student4":{"name":"David","marks":65,"grade":"C"}}
```

```python
nested_dict
```

```python
## itrate over these nestd dictinary

for outerkey,outervalue in nested_dict.items():
    print(f"outer key is: {outerkey}, outer value is: {outervalue}")
    for innerkey,innervalue in outervalue.items():
        print(f"inner key is: {innerkey}, inner value is: {innervalue}")
```

## Additional Functions in a lists are

### sum() - returns the sum of all the elements in the list
### min() - returns the minimum element in the list
### max() - returns the maximum element in the list
### len() - returns the number of elements in the list
### sorted() - returns a new list containing all elements from the original list in ascending order
### reversed() - returns a new list containing all elements from the original list in reverse order
### any() - returns True if any element of the list is true, otherwise returns False
### all() - returns True if all elements of the list are true, otherwise returns False
### enumerate() - returns an enumerate object that contains pairs of index and value for each element in the list
### zip() - returns an iterator of tuples, where the i-th tuple contains the i-th element from each of the argument sequences or iterables
### map() - applies a given function to each item of an iterable (like a list) and returns a list of the results
### filter() - constructs an iterator from elements of an iterable for which a function returns true
### reduce() - applies a rolling computation to sequential pairs of values in a list, reducing the list to a single value (requires functools module)
### list comprehension - a concise way to create lists using a single line of code, often with a for loop and an optional condition.
### Example: [x**2 for x in range(10) if x % 2 == 0] creates a list of squares of even numbers from 0 to 9.
### Note: The above functions are not exhaustive and there are many more functions available for lists in Python.

```python
## Sum()--> in list 
sum_values=sum(lists)
sum_values
```

```python
## min() and max() in list
min_value=min(lists)
max_value=max(lists)
print(f"Minimum value in the list: {min_value}")
print(f"Maximum value in the list: {max_value}")
```

```python
## len() in list
length_of_list=len(lists)
print(f"Length of the list: {length_of_list}")
```

```python
## sorted() in list
sorted_list=sorted(lists,reverse=True)
print(f"Sorted list in descending order: {sorted_list}")
```

```python
## reversed() in list
reversed_list=list(reversed(lists))
print(f"Reversed list: {reversed_list}")
```

```python
## any() and all() in list
any_value=any(lists)
all_value=all(lists)
print(f"Any value in the list is True: {any_value}")
print(f"All values in the list are True: {all_value}")
```

```python
## enumerate() in list --> enumerate() is a built-in function in Python that allows you to loop over a list (or any iterable) and have an automatic counter. It returns both the index and the value of each item in the list during the iteration.
for index, value in enumerate(lists):
    print(f"Index: {index}, Value: {value}")
```

```python
## zip() in list --> zip() is a built-in function in Python that takes multiple iterables (like lists) and aggregates them into a single iterable of tuples. Each tuple contains one element from each of the input iterables, paired together based on their positions.
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
zipped = list(zip(list1, list2))
print(f"Zipped lists: {zipped}")
```

```python
## map() in list --> map() is a built-in function in Python that applies a specified function to each item of an iterable (like a list) and returns an iterator that yields the results. It allows you to transform each element of the iterable according to the provided function.
def square(x):
    return x**2

mapped = list(map(square, lists))
print(f"Mapped list: {mapped}")
```

```python
## filter() in list --> filter() is a built-in function in Python that constructs an iterator from elements of an iterable for which a specified function returns true. It allows you to filter out elements from the iterable based on a condition defined by the provided function.
def is_even(x):
    return x % 2 == 0
filtered = list(filter(is_even, lists))
print(f"Filtered list (even numbers): {filtered}")
```

```python
## reduce() in list --> reduce() is a function from the functools module in Python that applies a specified binary function cumulatively to the items of an iterable, reducing the iterable to a single value. It takes two arguments: the function to apply and the iterable to reduce.
from functools import reduce
def add(x, y):
    return x + y
reduced = reduce(add, lists)
print(f"Reduced list: {reduced}")
```

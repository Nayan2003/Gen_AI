## 🔹 Sets in Python (English Notes + Code)

### 🔸 1. Definition

A **set** is an **unordered collection of unique elements**.

* No duplicates allowed
* No indexing (unordered)
* Mutable (can be modified)

---

## 🔸 2. Creating Sets

```python
# Using curly braces
s1 = {1, 2, 3, 4}

# Using set() function
s2 = set([1, 2, 2, 3])   # duplicates removed

print(s1)   # {1, 2, 3, 4}
print(s2)   # {1, 2, 3}
```

---

## 🔸 3. Adding Elements

```python
s = {1, 2, 3}

s.add(4)              # Add single element
s.update([5, 6, 7])   # Add multiple elements

print(s)
```

---

## 🔸 4. Removing Elements

```python
s = {1, 2, 3, 4}

s.remove(2)   # Removes element (error if not present)
s.discard(5)  # No error if element not present
s.pop()       # Removes a random element

print(s)
```

---

## 🔸 5. Set Operations

### ✅ Union (All elements from both sets)

```python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)
print(A.union(B))
```

---

### ✅ Intersection (Common elements)

```python
print(A & B)
print(A.intersection(B))
```

---

### ✅ Difference (Elements in A not in B)

```python
print(A - B)
```

---

### ✅ Symmetric Difference (Non-common elements)

```python
print(A ^ B)
```

---

## 🔸 6. Membership Test

```python
s = {1, 2, 3}

print(2 in s)   # True
print(5 in s)   # False
```

---

## 🔸 7. Set Methods

```python
s = {1, 2, 3}

print(len(s))     # Length of set
s.clear()         # Remove all elements
print(s)
```

---

## 🔸 8. Frozen Set (Immutable Set)

```python
fs = frozenset([1, 2, 3])

# fs.add(4) ❌ Not allowed (immutable)
print(fs)
```

---

## 🔥 Key Points (Exam Focus)

* Sets store **unique values only**
* Sets are **unordered**
* Useful for **removing duplicates**
* Support mathematical operations (union, intersection, etc.)

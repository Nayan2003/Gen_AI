## Palindrom string 

### Palindrome string is going to have characcter similarto it from both the sides or even if you wwwell reverse the string still you will be getting the gives string itself 


## palindrome --> aba, aaabbbaaa,zzyyzz

## Not a Palindrome --> ab, abas , Aba != abA ,,, ab a != a ba

```python
## Checckk that wwheter if the current string is a palindrome or not


def palindrom(string):
    string_value=string.lower().replace(" ","")

    if string_value == string_value[::-1]:
        print(f"The urrent string is  ({string}) a pallindrom")

    else:
        print(f"given string is ( {string} ) not pallindrome")
```

```python
palindrom("Naya n")
```

```python
palindrom("ab") # ab != ba
```

```python
palindrom("AaBB aa") # aabbaa
```

```python
## Calculate the total cost of items in a shopping ccart

cart = [{'item1':'Apple','cost':5,'Quantity':20},
        {'item2':'Banana','cost':4,'Quantity':15},
        {'item3':'Guava','cost':3,'Quantity':10}]


# indexx in our cart list ---> Keyname
## Total --> Cost of 1item * Quantity


def cart_total(cart_item):
    total_cost=0

    for item in cart_item:
        total_cost+=item['cost'] * item['Quantity']

    print("The Total cost for the cart items is going to be :", total_cost)
```

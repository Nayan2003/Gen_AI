# Most Important Parts in Recursion

Every recursive function has 2 things:

---

## 1. Base Case (Stopping Condition)

```python
if num == 0 or num == 1:
    return 1
```

This stops recursion.

Otherwise function would call itself forever.

---

## 2. Recursive Call

```python
return num * fac_recursive(num-1)
```

Function calls itself with smaller value.

---

# Step-by-Step Flow

Suppose:

```python
fac_recursive(4)
```

---

## Step 1

```python
fac_recursive(4)
```

Not base case.

So:

```python
return 4 * fac_recursive(3)
```

Python waits for value of `fac_recursive(3)`.

---

## Step 2

```python
fac_recursive(3)
```

returns:

```python
3 * fac_recursive(2)
```

---

## Step 3

```python
fac_recursive(2)
```

returns:

```python
2 * fac_recursive(1)
```

---

## Step 4 (Base Case)

```python
fac_recursive(1)
```

returns:

```python
1
```

Now recursion stops.

---

# Now values go BACK upward

This is the most important concept.

Python now solves:

---

From Step 3:

```python
2 * 1 = 2
```

So:

```python
fac_recursive(2) = 2
```

---

From Step 2:

```python
3 * 2 = 6
```

So:

```python
fac_recursive(3) = 6
```

---

From Step 1:

```python
4 * 6 = 24
```

So:

```python
fac_recursive(4) = 24
```

---

# Final Output

```python
print(fac_recursive(4))
```

Output:

```python
24
```

# Visual Recursive Tree

```python
fac_recursive(4)

```

Then after reaching base case:

```python

```

comes back upward.

# Functions in Python

A **function** is a reusable block of code that performs a specific task.

Instead of writing the same code again and again, we place it inside a function and call it whenever needed.

---

# 1. Function with a Single Parameter

A function that accepts only one value/input.

---

## Syntax

```python
def function_name(parameter):
    # code
```

---

## Example

```python
def square(num):
    return num * num

print(square(5))
```

Output:

```python
25
```

---

## Explanation

Here:

```python
num
```

is the parameter.

When:

```python
square(5)
```

is called,
`5` gets stored inside `num`.

Then:

```python
5 * 5 = 25
```

---

# 2. Function with Multiple Parameters

A function that accepts more than one input.

---

## Syntax

```python
def function_name(param1, param2):
    # code
```

---

## Example

```python
def add(a, b):
    return a + b

print(add(10, 20))
```

Output:

```python
30
```

---

## Explanation

Here:

* `a = 10`
* `b = 20`

Then:

```python
10 + 20 = 30
```

# 3. Function with Default Parameters

Default parameter means:

> If user does not provide value, Python uses default value automatically.

---

## Syntax

```python
def function_name(parameter=value):
```

---

## Example

```python
def greet(name="Guest"):
    print(f"Hello {name}")

greet()
greet("Nayan")
```

Output:

```python
Hello Guest
Hello Nayan
```

---

## Explanation

### First call

```python
greet()
```

No argument passed.

So default value is used:

```python
name = "Guest"
```

---

### Second call

```python
greet("Nayan")
```

Now default value gets replaced.

```python
name = "Nayan"
```

# 4. Function with Variable Length Parameters

Sometimes we don't know how many inputs user will provide.

Python provides:

* `*args` → positional arguments
* `**kwargs` → keyword arguments

---

# A) Variable Length Positional Arguments (`*args`)

Used when number of arguments is unknown.

`*args` stores values in the form of a tuple.

---

## Syntax

```python
def function_name(*args):
```

---

## Example

```python
def numbers(*args):
    print(args)

numbers(1, 2, 3, 4)
```

Output:

```python
(1, 2, 3, 4)
```

---

## Explanation

All values are collected into one tuple.

```python
args = (1,2,3,4)
```

---

## Another Example

```python
def addition(*nums):
    total = 0

    for i in nums:
        total += i

    return total

print(addition(10, 20))
print(addition(1, 2, 3, 4, 5))
```

Output:

```python
30
15
```

---

## Internal Working

When:

```python
addition(1,2,3)
```

Python converts it into:

```python
nums = (1,2,3)
```

Then loop runs through tuple.

---

# B) Variable Length Keyword Arguments (`**kwargs`)

Used when arguments are passed with names (key=value).

`**kwargs` stores data in dictionary form.

---

## Syntax

```python
def function_name(**kwargs):
```

---

## Example

```python
def student(**kwargs):
    print(kwargs)

student(name="Nayan", age=22)
```

Output:

```python
{'name': 'Nayan', 'age': 22}
```

---

## Explanation

Python stores keyword arguments as dictionary.

```python
kwargs = {
    "name": "Nayan",
    "age": 22
}
```

---

## Loop Example

```python
def student(**kwargs):

    for key, value in kwargs.items():
        print(key, ":", value)

student(name="Nayan", age=22, city="Mumbai")
```

Output:

```python
name : Nayan
age : 22
city : Mumbai
```

---

# Difference Between `*args` and `**kwargs`

| Feature        | `*args`             | `**kwargs`                  |
| -------------- | ------------------- | --------------------------- |
| Full Form      | Arbitrary Arguments | Arbitrary Keyword Arguments |
| Stores Data As | Tuple               | Dictionary                  |
| Input Type     | Positional          | Key-value                   |
| Symbol         | `*`                 | `**`                        |

---

# Combined Example

```python
def demo(a, *args, **kwargs):

    print("Normal Parameter:", a)
    print("Positional:", args)
    print("Keyword:", kwargs)

demo(10, 20, 30, name="Nayan", city="Mumbai")
```

Output:

```python
Normal Parameter: 10
Positional: (20, 30)
Keyword: {'name': 'Nayan', 'city': 'Mumbai'}
```

# Parameter Flow in Python

Python accepts parameters in this order:

1. Normal parameters
2. Default parameters
3. `*args`
4. `**kwargs`

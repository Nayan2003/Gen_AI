Here are **detailed Python notes on Functions** (exam + practical friendly). I’ll cover concept, syntax, types, arguments, scope, recursion, lambda, decorators (basic), and best practices—with examples.

---

## 1) What is a Function in Python?

A **function** is a reusable block of code that performs a specific task.

### Why we use functions

* **Reusability** (same logic many times)
* **Modularity** (break big program into small parts)
* **Readability**
* **Easy debugging & testing**

---

## 2) Function Syntax

```python
def function_name(parameters):
    # body
    return value
```

### Example

```python
def add(a, b):
    return a + b

print(add(5, 3))  # 8
```

---

## 3) Parts of a Function

### (A) `def` keyword

Defines a function.

### (B) Function name

Should be meaningful, like `calculate_total`, `get_user_age`.

### (C) Parameters

Inputs to the function (optional).

### (D) Function body

Indented code block.

### (E) `return` statement

Sends output back to caller (optional).
If no return → returns `None`.

---

## 4) Calling a Function

```python
def greet():
    print("Hello!")

greet()  # function call
```

---

## 5) Return vs Print

### Print

Shows output on screen, doesn’t “give back” value.

```python
def f():
    print(10)

x = f()
print(x)  # None
```

### Return

Gives value to the caller.

```python
def f():
    return 10

x = f()
print(x)  # 10
```

---

## 6) Types of Functions

### 1) Built-in Functions

Already provided by Python.
Examples: `len()`, `sum()`, `type()`, `range()`, `max()`

### 2) User-defined Functions

Created by us using `def`.

### 3) Anonymous Functions (Lambda)

Short one-line function using `lambda`.

---

## 7) Function Arguments & Parameters (Most Important)

### Parameter vs Argument

* **Parameter**: variable in function definition
* **Argument**: actual value passed when calling

```python
def add(a, b):   # a, b are parameters
    return a+b

add(2, 3)        # 2, 3 are arguments
```

---

# 8) Types of Arguments

## (A) Positional Arguments

Order matters.

```python
def info(name, age):
    print(name, age)

info("Nayan", 22)   # correct
```

---

## (B) Keyword Arguments

Pass by parameter name (order not required).

```python
info(age=22, name="Nayan")
```

---

## (C) Default Arguments

If value not given, default is used.

```python
def greet(name="User"):
    print("Hello", name)

greet()           # Hello User
greet("Nayan")    # Hello Nayan
```

⚠️ **Default argument caution (mutable default)**
Bad:

```python
def add_item(item, lst=[]):
    lst.append(item)
    return lst
```

Better:

```python
def add_item(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst
```

---

## (D) Variable Length Arguments

When you don’t know how many values will come.

### 1) `*args` (many positional)

```python
def total(*args):
    return sum(args)

print(total(1,2,3,4))  # 10
```

### 2) `**kwargs` (many keyword)

```python
def profile(**kwargs):
    for k, v in kwargs.items():
        print(k, "=", v)

profile(name="Nayan", age=22, city="Mumbai")
```

### Mixing all types (order is important)

✅ Correct order in function definition:

```python
def fun(a, b=10, *args, **kwargs):
    pass
```

---

## 9) Multiple Return Values

Python returns them as a tuple.

```python
def calc(a, b):
    return a+b, a-b, a*b

x = calc(10, 5)
print(x)  # (15, 5, 50)
```

Unpacking:

```python
s, d, m = calc(10, 5)
```

## 10) Scope of Variables (Local vs Global)

### Local variable

Inside function only.

```python
def f():
    x = 10
    print(x)

f()
# print(x)  # error
```

### Global variable

Outside function.

```python
x = 100
def f():
    print(x)

f()  # 100
```

### Modify global inside function: `global`

```python
x = 5
def f():
    global x
    x = x + 1

f()
print(x)  # 6
```

## 13) Recursion (Function calls itself)

Used in problems like factorial, fibonacci, tree traversal.

### Factorial example

```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n-1)

print(fact(5))  # 120
```

⚠️ Must have **base condition** to stop recursion.

---

## 14) Lambda (Anonymous Function)

One-line function without `def`.

```python
square = lambda x: x*x
print(square(6))  # 36
```

Common with `map`, `filter`, `sorted`.

### map

```python
nums = [1,2,3]
print(list(map(lambda x: x*2, nums)))  # [2,4,6]
```

### filter

```python
nums = [1,2,3,4,5]
print(list(filter(lambda x: x%2==0, nums)))  # [2,4]
```

### sorted with key

```python
names = ["bbb", "a", "cccc"]
print(sorted(names, key=lambda x: len(x)))
```

---

## 15) Higher-Order Functions

Function that:

* takes another function as input, or
* returns a function

Example:

```python
def apply(f, x):
    return f(x)

print(apply(lambda n: n*n, 5))  # 25
```

---

## 16) Nested Functions

Function inside another function.

```python
def outer():
    def inner():
        return "inner"
    return inner()

print(outer())
```

---

## 17) Closures (advanced but useful)

Inner function remembers outer function variables.

```python
def multiplier(k):
    def mul(x):
        return x * k
    return mul

times3 = multiplier(3)
print(times3(10))  # 30
```

---

## 18) Decorators (basic concept)

A decorator adds extra behavior to a function.

```python
def deco(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@deco
def hello():
    print("Hello!")

hello()
```

Output:

```
Before
Hello!
After
```

---

## 19) Best Practices for Functions

* Keep functions **small and specific**
* Use **meaningful names**
* Prefer **return** instead of only print
* Use **docstrings**
* Avoid too many global variables
* Handle errors using try/except when needed

---

## 20) Quick Summary Sheet (Exam-ready)

* `def` defines function
* `return` sends value (no return → `None`)
* Arguments: positional, keyword, default, `*args`, `**kwargs`
* Scope: local, global (`global` keyword)
* Recursion needs base condition
* Lambda: small one-line function
* Decorator: function that modifies another function
